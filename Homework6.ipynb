```python
#Work4
ans = input().split()
result = {}
for x in ans:
    key, value = x.split(":")
    if key in result:
        result[key].append(value)
    else:
        result[key] = [value]
print(result)
```

```python
#Work5
numbers = map(int, input().split())
a = 0
b = 0
c = 0
ans = {"positives":a, "negatives":b, "zeros":c}
for number in numbers:
    if number > 0:
        a += 1
        ans["positives"] = a
    elif number < 0:
        b +=1
        ans["negatives"] = b
    else:
        c +=1
        ans["zeros"] = c
print(ans)
```

```python
#Work7
line = tuple(map(int, input().split()))
a = int(line[0])
b = int(line[-1])
def reverse_line(line):
    ans = []
    for i in range(len(line)-1, -1, -1):
        x = str(line[i])
        ans.append(x)
    return tuple(ans)
print(a,b,reverse_line(line))
```

```python
#Work9
d1 = eval(input())
d2 = eval(input())
for k, v in d2.items():
    if k in d1:
        d1[k] = d1[k] + v
    else:
        d1[k] = v
ans = dict(sorted(d1.items()))
print(ans)
```

```python
#Work10
numbers = list(map(int, input().split()))
k = int(input())
ans1 = []
for i in range(len(numbers)):
    for j in range(i+1, len(numbers)):
        if numbers[i] + numbers[j] == k:
            pairs = (numbers[i], numbers[j])
            ans1.append(pairs)
ans1.sort()
print(ans1)
```

```python
#Work12
numbers = input().split()
ans = {}
for v in numbers:
    if v in ans:
        ans[v] +=1
    else:
        ans[v] = 1
max_count = 0
best_x = numbers[0]
for v, c in ans.items():
    if max_count < c:
        max_count = c
        best_x = v
    if max_count == c:
        if best_x > v:
            best_x = v
print(best_x)
```

```python
#Work15
line = input().split()
k = int(input())
ans = {}
result = {}
for i in range(0, len(line), 2):
    key = line[i]
    v = int(line[i+1])
    ans[key] = v
for key, v in ans.items():
    if v > k:
        result[key] = v
print(result)
```

```python
#Work16
n, m = map(int, input().split())
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    if len(row) != m:
        print("Error")
    matrix.append(row)
for row in matrix:
    for v in row:
        print(f"{v:>4}",end = "")
    print()
```

```python
#Work18
n, m, k = map(int, input().split())
ans = []
counta = 0
for i in range(n):
    if i == k-1:
        row = list(map(int, input().split()))
        ans.append(row[k-1])
        countb = 0
        for v in row:
            countb += v
    else:
        row = list(map(int, input().split()))
        ans.append(row[k-1])
for x in ans:
    counta += x
print(counta, countb)
```
