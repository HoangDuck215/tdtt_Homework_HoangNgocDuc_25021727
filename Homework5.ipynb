```python
#Work1
def Max_Of_Two(a, b):
    max = a
    if max < b:
        max = b
    return max
a, b = map(int, input().strip().split())
print(Max_Of_Two(a, b))
```

```python
#Work2
def swap(a, b):
    c = a
    a = b
    b= c
    return a, b
a, b = map(int, input().split())
a, b = swap(a, b)
print(a, b)
```

```python
#Work3
def prime(x):
    if x == 1 or x ==0:
        return False
    if x > 0:
        for i in range(2, x):
            if x % i ==0:
                return False
        return True
    else:
        return False
n = int(input())
print(prime(n))
```

```python
#Work4
def perfect(x):
    sum = 0
    for i in range(1, x):
        if x % i == 0:
            sum +=i
    if sum == x:
        return True
    else:
        return False
n = int(input())
print(perfect(n))
```

```python
#Work5
def position(x):
    ans = 0
    for i in range(0, len(lst)):
        if lst[i] == x:
            ans = i+1
            break
        if lst[i] != x:
            ans = -1
    return ans
lst = list(map(int, input().split()))
n = int(input())
print(position(n))
```

```python
#Work6
def factorial(x):
    ans = 1
    for i in range(1, x+1):
        ans *= i
    return ans
n = int(input())
print(factorial(n))
```

```python
#Work7
def cal(s):
    for i in range(0, len(s)):
        if s[i] == "+":
            a = float(s[0:i])
            b = float(s[i+1:len(s)])
            ans = a+b
        if s[i] == "-":
            a = float(s[0:i])
            b = float(s[i+1:len(s)])
            ans = a-b
        if s[i] == "*":
            a = float(s[0:i])
            b = float(s[i+1:len(s)])
            ans = a*b
        if s[i] == "/":
            a = float(s[0:i])
            b = float(s[i+1:len(s)])
            ans = a/b
    return ans
n = input()
print(cal(n))
```

```python
#Work8
def hamming(a, b):
    XOR = a ^ b
    ans = bin(a ^ b)
    return ans.count("1")
a, b = map(int, input().split())
print(hamming(a, b))
```

```python
#Work9
def sum(x):
    ans = 0
    while x > 0:
        a = x % 10
        x = x //10
        ans += a
    return ans
n = int(input())
print(sum(n))
```

```python
#Work10
def isomorphism(a, b):
    x = {}
    y = {}
    for char_a, char_b in zip(a, b):
        if char_a in x:
            if x[char_a] != char_b:
                return False
        else:
            x[char_a] = char_b
        if char_b in y:
            if y[char_b] != char_a:
                return False
        else:
            x[char_b] = char_a
    return True
a, b = map(str, input().split())
print(isomorphism(a, b))
```
