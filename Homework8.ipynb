```python
#Work1
class cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
    def total_area(self):
        pi = 3.14
        area = 2*(pi*self.radius) + 2*pi*self.radius*self.height
        return area
    def volume(self):
        pi = 3.14
        tt =  pi*(self.radius**2)
        return tt
r, h = map(float, input().split())
hinh_tru = cylinder(r, h)
print(round(hinh_tru.total_area(), 2))
print(round(hinh_tru.volume(), 2))
```

```python
#Work2
class date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    def next_day(self):
        days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.year % 400 == 0 or (self.year % 4 ==0 and self.year %100!= 0):
            days[2] = 29
        if self.month > 12:
            return "INVALID"
        if self.day > days[self.month]:
            return "INVALID"
        self.day +=1
        if self.day > days[self.month]:
            self.day = 1
            self.month +=1
            if self.month > 12:
                self.month = 1
                self.year +=1
        return f"{self.day}/{self.month}/{self.year}"
lst = input().split("/")
dates = list(map(int, lst))
next_date = date(dates[0], dates[1], dates[2])
print(next_date.next_day())
```

```python
#Work5
class shoppingcart:
    def __init__(self):
        self.cart = []
    def add(self, name, price):
        item = {"name":name, "price":price}
        self.cart.append(item)
        print(f"Đã thêm {name} vào giỏ hàng.")
    def delete(self, name):
        for item in self.cart:
            if item["name"] == name:
                self.cart.remove(item)
            else:
                print("Không tìm thấy sản phẩm!")
        print(f"Đã xoá {name} thành công.")
    def cart_check(self):
        if self.cart == []:
            print("Giỏ hàng trống.")
        else:
            print("Giỏ hàng có sản phẩm!")
    def check(self):
        count = 0
        for item in self.cart:
            count += item["price"]
        print(f"Đây là tổng tiền: {count}")
    def cart_show(self):
        if self.cart == []:
            print("Giỏ hàng trống!")
        else:
            for item in self.cart:
                print(f"Tên: {item['name']}/Giá: {item['price']}")
    def delete_all(self):
        self.cart.clear()
        print("Đã xoá hết sản phẩm!")
my_cart = shoppingcart()
while True:
    print("=====SHOPPING CART MENU=====")
    print("2.Thêm vào giỏ hàng một sản phẩm mới")
    print("3.Xoá khỏi giỏ hàng một sản phẩm mới")
    print("4.Kiểm tra xem giỏ hàng có rỗng không")
    print("5.Tính tổng tiền giỏ hàng")
    print("6.Hiển thị toàn bộ sản phẩm trong giỏ")
    print("7.Xoá toàn bộ giỏ hàng")
    n = int(input())
    if n == 2:
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá tiền sản phẩm: "))
        my_cart.add(name, price)
    elif n == 3:
        delete_name = input("Nhập tên sản phẩm muốn xoá: ")
        my_cart.delete(delete_name)
    elif n == 4:
        my_cart.cart_check()
    elif n == 5:
        my_cart.check()
    elif n ==6:
        my_cart.cart_show()
    elif n ==7:
        my_cart.delete_all()
    ans = input("Có muốn thoát khỏi chương trình không ?: ")
    if ans.lower() == "yes":
        break
```

```python
#Work6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return 2*(self.w + self.h)
    def scale(self, k):
        self.w = k*self.w
        self.h = k*self.h
w, h, k  = map(int, input().split())
hcn = Rectangle(w, h)
hcn.scale(k)
print(hcn.area())
print(hcn.perimeter())
```

```python
#Work7
import math
class Fraction:
    def __init__(self, num, den):
        if den <0:
            num = -num
            den = -den
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common
    def add(self, other):
        new_num = self.num * other.den + self.den * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    def sub(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    def mul(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    def div(self, other):
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)
    def __str__(self):
        return f"{self.num}/{self.den}"
s = input().split()
a = int(s[0])
b = int(s[1])
op = s[2]
c = int(s[3])
d = int(s[4])
phanso1 = Fraction(a, b)
phanso2 = Fraction(c, d)
if op == "+":
    ans = phanso1.add(phanso2)
elif op == "-":
    ans = phanso1.sub(phanso2)
elif op == "*":
    ans = phanso1.mul(phanso2)
elif op == "/":
    ans = phanso1.div(phanso2)
print(ans)
```

```python
#Work11
import math
class vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def sub(self, other):
        return vector(self.x-other.x, self.y - other.y, self.z-other.z)
    def tichcohuong(self, other):
        return vector(self.y*other.z-self.z*other.y, self.z*other.x-self.x*other.z, self.x*other.y- self.y*other.x)
    def tichvohuong(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    def length(self):
        return (self.x**2 + self.y**2 + self.z**2)**(1/2)
n = list(map(float, input().split()))
A = vector(n[0], n[1], n[2])
m = list(map(float, input().split()))
B = vector(m[0], m[1], m[2])
i = list(map(float, input().split()))
C = vector(i[0], i[1], i[2])
j = list(map(float, input().split()))
D = vector(j[0], j[1], j[2])
vector_AB = A.sub(B)
vector_BC = B.sub(C)
vector_CD = C.sub(D)
v1 = vector_AB.tichcohuong(vector_BC)
v2 = vector_BC.tichcohuong(vector_CD)
s = v1.tichvohuong(v2)
v1_length = v1.length()
v2_length = v2.length()
if v1_length*v2_length == 0:
    print("ERROR")
else:
    cos_phi = (abs(s))/(v1_length*v2_length)
    if cos_phi > 1.0:
        cos_phi = 1.0
    if cos_phi < -1.0:
        cos_phi = -1.0
    angle = math.degrees(math.acos(cos_phi))
    print(f"{angle:.2f}")
```

```python
#Work12
class Complex:
    def __init__(self, re, im):
        self.re = re
        self.im = im
    def mul(self, other):
        return Complex(round(self.re*other.re - self.im*other.im), round(self.re*other.im + self.im*other.re))
    def __str__(self):
        return f"{self.re} {self.im}"
a, b = map(float, input().split())
c, d = map(float, input().split())
if abs(a) > 100 or abs(b) > 100 or abs(c) > 100 or abs(d) > 100:
    print("Error")
else:
    c1 = Complex(a, b)
    c2 = Complex(c, d)
    print(c1.mul(c2))
```

```python
#Work13
class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3
    def Area(self):
        if self.edge1 + self.edge2 <= self.edge3 or self.edge1 + self.edge3 <= self.edge2 or self.edge3 + self.edge2 <= self.edge1:
            return -1
        else:
            p = (self.edge1 + self.edge2 + self.edge3)/2
            return round((p*(p-self.edge1)*(p-self.edge2)*(p-self.edge3))**(1/2),2)
a, b, c = map(int, input().split())
tamgiac = Triangle(a, b, c)
result = tamgiac.Area()
if result == -1:
    print("Invalid")
else:
    print(f"{result:.2f}")
```
