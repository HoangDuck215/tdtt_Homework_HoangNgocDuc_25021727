```python
#Work1
def binary_search(arr, left, right, target):
    a = arr
    L = left
    R = right
    while R >= L:
        mid = L + (R-L)//2
        if target == a[mid]:
            return mid
        elif target > a[mid]:
            L = mid +1
        elif target < a[mid]:
            R = mid - 1
    return -1
arr = list(map(int , input().split()))
left = 0
right = len(arr)-1
target = int(input())
print(binary_search(arr, left, right, target))
```

```python
#Work2
def count_occurrences(arr, x) -> int:
    count = 0
    for v in arr:
        if v == x:
            count +=1
    return count
arr = list(map(int, input().split()))
x = int(input())
print(count_occurrences(arr, x))
```

```python
#Work3A
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0, len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
arr = list(map(int, input().split()))
print(bubble_sort(arr))
```

```python
#Work3B
def selection_sort(arr):
    for i in range(len(arr)):
        min = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min]:
                min = j
        arr[i], arr[min] = arr[min], arr[i]
    return arr
arr = list(map(int, input().split()))
print(selection_sort(arr))
```

```python
#Work4
def most(arr):
    max = 0
    for v in arr:
        count = arr.count(v)
        if max < count:
            max = count
            best_x = v
        if max == count:
            break
    print(f"{best_x} xuat hien nhieu nhat, som nhat,{count} lan")
arr = list(map(int, input().split()))
most(arr)
```
