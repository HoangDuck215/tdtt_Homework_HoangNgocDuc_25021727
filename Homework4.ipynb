```python
#Work8
def reverse(n):
    x = 0
    while n !=0:
        a = n%10
        x = x*10+a
        n = n//10
    return x
n = int(input())
count = 0
while n != reverse(n):
    n = n + reverse(n)
    count +=1
print(count, n)
```

```python
#Work9
n = int(input())
i = 0
result =[]
while i*i <=n:
    square = i*i
    square_str=str(square)
    if len(square_str) == len(set(square_str)):
        result.append(square)
    i+=1
if len(result) >0:
    print(" ".join(map(str, result)))
else:
    print("(no number)")
```

```python
#Work10
n = int(input())
def collatz(x):
    count = 1
    while x !=1:
        if x % 2 ==0:
            x = x/2
        else:
            x = 3*x+1
        count +=1
    return count
max_count = 1
best_x = 1
for x in range(1, n+1):
    current_count = collatz(x)
    if current_count > max_count:
        max_count = current_count
        best_x = x
print(best_x, max_count)
```
